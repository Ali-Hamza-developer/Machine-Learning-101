# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read Data

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head(5)
# we only predicted based on Age Purchased or not based on Age
```

```python
df.drop(columns=["EstimatedSalary"], inplace=True)
```

```python
df.head(5)
```

# Plot 


```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age", y="Purchased", data=df)
plt.show()
```

```python
x = df[["Age"]]
y = df["Purchased"]
```

# Train_Test_Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test , y_train,  y_test =train_test_split(x, y, test_size = 0.2, random_state = 42)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

# Train

```python
lr.fit(x_train, y_train)
```

# Accuracy

```python
lr.score(x_test, y_test)*100
```

# PLot Look like

```python
y_pred = lr.predict(x)
```

```python
plt.figure(figsize=(5,4))

# scatterplot of original data
sns.scatterplot(x="Age", y="Purchased", data=df)

# lineplot of predictions
sns.lineplot(x=df["Age"], y=y_pred, color="red")

plt.show()
```
